````python
data = boston['data']
target = boston['target']
features = boston['feature_names']
descr = boston['DESCR']
filename = boston['filename']
````

### Create a dataframe with the array ```features``` and the correct column names

````python
features
````

````python
boston_house = pd.DataFrame(data, columns = features)
````

````python
boston_house
````

### Add a MEDV column to your dataframe using the values from target

````python
mdev_df = pd.DataFrame(target,columns=['MEDV'])
````

````python
mdev_df
````

````python
boston_house_with_medv = pd.concat([boston_house, mdev_df], axis=1)
````

### Add a column called ONES to the dataframe which consists of all 1s

````python
boston_house_with_medv['ONES'] = 1
````

### Add a column to the dataframe called MEDV_times_TAX which is the product of MEDV and TAX

````python
boston_house_with_medv['MEDV_times_TAX'] = boston_house_with_medv['MEDV'] * boston_house_with_medv['TAX']
````

### What is the average MEDV of houses located on the Charles River
